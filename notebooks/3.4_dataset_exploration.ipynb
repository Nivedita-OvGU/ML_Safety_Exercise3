```python
# unzip the train and test files
from pathlib import Path
import zipfile

PROJECT_ROOT = Path("..").resolve()
RAW_DIR = PROJECT_ROOT / "data" / "raw"
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

TRAIN_ZIP = RAW_DIR / "train.zip"
TEST_ZIP = RAW_DIR / "test.zip"

TRAIN_DIR = PROCESSED_DIR / "train"
TEST_DIR = PROCESSED_DIR / "test"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

def unzip_file(zip_path, output_dir):
    output_dir.mkdir(parents=True, exist_ok=True)
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(output_dir)
    print(f"Extracted: {zip_path.name} → {output_dir}")

unzip_file(TRAIN_ZIP, TRAIN_DIR)
unzip_file(TEST_ZIP, TEST_DIR)

```

```python
#check what's inside the dataset
def show_folder_tree(path, max_depth=3):
    path = Path(path)
    for item in sorted(path.rglob("*")):
        depth = len(item.relative_to(path).parts)
        if depth <= max_depth:
            print("  " * depth + item.name)

print("TRAIN FOLDER:")
show_folder_tree(TRAIN_DIR, max_depth=3)

print("\nTEST FOLDER:")
show_folder_tree(TEST_DIR, max_depth=3)
```

```python
# to find image and label files automatically
IMAGE_EXTENSIONS = [".jpg", ".jpeg", ".png", ".bmp"]

def find_images(folder):
    images = []
    for ext in IMAGE_EXTENSIONS:
        images.extend(Path(folder).rglob(f"*{ext}"))
    return sorted(images)

def find_csv_files(folder):
    return sorted(Path(folder).rglob("*.csv"))

train_images = find_images(TRAIN_DIR)
test_images = find_images(TEST_DIR)

train_csvs = find_csv_files(TRAIN_DIR)
test_csvs = find_csv_files(TEST_DIR)

print("Number of training images:", len(train_images))
print("Number of test images:", len(test_images))

print("\nCSV files in train:")
for file in train_csvs:
    print(file)

print("\nCSV files in test:")
for file in test_csvs:
    print(file)
```

```python
# load the label CSV. Exercise 3.4.1
# How many images are in the training and test splits?
import pandas as pd

train_labels_path = train_csvs[0]
test_labels_path = test_csvs[0]

train_df = pd.read_csv(train_labels_path)
test_df = pd.read_csv(test_labels_path)

print("Train labels shape:", train_df.shape)
print("Test labels shape:", test_df.shape)

print("\nTrain columns:")
print(train_df.columns.tolist())

display(train_df.head())
```
